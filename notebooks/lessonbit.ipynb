# AP CSP 4.1: Bit Rate, Bandwidth, & Latency

**Three essential networking concepts:**
- **Bit Rate**: How fast data moves (bits/second)
- **Bandwidth**: Maximum capacity available (bits/second)  
- **Latency**: Time delay in transmission (milliseconds)

## Vocabulary

**Bit Rate** - Data transmitted per second (bps, Mbps, Gbps)  
**Bandwidth** - Maximum capacity of a connection  
**Latency** - Delay between sending/receiving data (milliseconds)  
**Throughput** - Actual data successfully transmitted (often less than bandwidth)

## 1. Bit Rate = Speed of Data Transmission

**Formula:** Bit Rate (bps) = Number of Bits / Time (seconds)

**Example:** 1,000,000 bits in 1 second = 1 Mbps

**Common speeds:** 56 Kbps (dial-up) → 100 Mbps (cable) → 1+ Gbps (fiber)

## 2. Bandwidth = Maximum Capacity

Think of a water pipe—bandwidth is the pipe size; bit rate is how much flows through.

**Key difference:** 
- Bandwidth = theoretical maximum
- Throughput = actual measured speed (always ≤ bandwidth)

**Example:** ISP promises 300 Mbps (bandwidth), but you get 240 Mbps (throughput)

## 3. Latency = Time Delay (Response Time)

Caused by: physical distance, network congestion, routing delays

**Acceptable latency:**
- Web browsing: < 100 ms
- Gaming: < 50 ms  
- Video streaming: < 200 ms

**Critical insight:** High bandwidth ≠ Low latency. They're independent!
- Satellite internet: HIGH bandwidth, VERY HIGH latency (500+ ms)

## Quick Practice

**Q1: Download time?** 500 MB file at 50 Mbps = ?
- 500 MB = 4,000 Mb ÷ 50 Mbps = **80 seconds**

**Q2: Which matters most?**
- Large file download → **Bandwidth** (speed)
- Online gaming → **Latency** (responsiveness)

## Key Takeaway

| Concept | Measures | Unit | Impact |
|---------|----------|------|--------|
| **Bit Rate** | Speed | bps | Download speed |

| **Latency** | Delay | ms | Responsiveness |
